```python
#Task 1
def grading_system():
    while True:
        score = input("Enter score (0-100) or type 'exit' to quit the program: ")
        if score.lower() == 'exit':
            print("Exiting grading system.")
            break
        if not score.isdigit():
            print("Invalid input. Please enter a numeric value")
            continue

        score = int(score)
        if score < 0 or score > 100:
            print("score must be between 0 and 100")
            continue

        if score >=90:
            grade = "A"
        elif score >=80:
            grade = "B"
        elif score >=70:
            grade = "C"
        elif score >=60:
            grade = "D"
        else:
            grade = "F"

        print(f"Grade: {grade}")

#call the function
grading_system()
```
