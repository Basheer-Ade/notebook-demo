```python
#Task2.1
print("Even Number Printer")
print ("Even numbers from 1 to 100:")

for num in range(1,101):
    if num % 2 == 0:
        print(num)        
```

```python
#Task2.2
print ("Numbers between 1-50 that are divisible by both 3 and 5:")

for numbers in range(1,51):
    if numbers % 3 == 0 and numbers % 5 ==0:
        print(numbers)
```

```python
#Task2.3
print("Custom Menu App")
def custom_menu_app():
    while True:
        print("\nMenu:")
        print("1. Greet\n" "2. Add\n" "3. Exit")
        name = input("Enter your name")
        option = input("Enter your option: ")
        
        if option == "1":
            print(f" Hello, {name} !")
        elif option =="2":
            try:
                a = float(input("Enter the first number:"))
                b = float(input("Enter the second number:"))
                print(f" The sum of the two numbers is {a + b}")
            except ValueError:
                print(f" {name}, please enter valid numbers.")
        elif option == "3":
            print (f" Goodbye, {name}!")
            break
        else:
            print(f"{name}, please enter a valid option. Try again.")


# Call the function
custom_menu_app()
   
    
```
