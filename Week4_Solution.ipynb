```python
# Function 1: Calculate_discount

def calculate_discount(price, rate):
    """
    Calculates the discounted price given the original price and discount rate.
    
    Parameters:
    price(float): The original price of the item
    rate(float): The discount rate (e.g., 0.1 for 10%)
    
    Returns:
    float: The price after applying the discount.
    """
    return price - (price * rate)

#Test cases for calculate_discount
print("Discounted price 1: ", calculate_discount(100, 0.1))
print("Discounted price 2: ", calculate_discount(250, 0.25))
```

```python
#Function 2: Format_name

def format_name(first, last):
    """
    Formats a full name with proper capitlization.

    Parameters:
    first (str): First name
    last (str): Last name

    Returns:
    str: Formatted full name (e.g., 'John Doe')
    """
    
    return f"{first.strip().capitalize()} {last.strip().capitalize()}"
    
#Test cases for format_name
print ("Formatted Name1:", format_name("John", "Doe"))
print ("Formatted Name2:", format_name("Alice", "Smith"))
```

```python
#Function 3: is_valid_email

def is_valid_email(email):
    """
    Validates whether an email conatin '@' and '.com'.

    Parameters:
    email (str): Email address to check

    Returns:
    bool: True if email is valid, false otherwise
    """
    
    return '@' in email and '.com' in email
    
#Test cases for is_valid_email
print ("Email Valid 1:", is_valid_email("user@example.com"))
print ("Email Valid 2:", is_valid_email("invalid-email@domain"))
```

```python
#Function 4: mini_application

def mini_application():
    """
    A small application that:
    -Validates an email
    -Formats the user's full name
    -Calculates and displays a discounted price

    Inputs are taken directly from a user.
    """
    print("Welcome to the Customer Utility App!")

    #Input: Name
    first = input("Enter your first name: ")
    last = input("Enter your last name: ")
    formatted = format_name(first, last)

    #Input: Email
    email = input("Enter your email address: ")
    if is_valid_email(email):
        print("Email is valid")
    else:
        print("Invalid email format.")
        return #Exit the fucntion early if the email is invalid

    #Input: Price and discount
    try:
        price = float(input("Enter the original price in $"))
        discount_rate = float(input("Enter discount rate(e.g., 0.2 for 20%: "))
        final_price = calculate_discount(price, discount_rate)
        print(f"\nHello, {formatted}!")
        print(f"Discounted price: ${final_price:.2f}")
    except ValueError:
        print("Invalid number entered for price or discount")
        
```
