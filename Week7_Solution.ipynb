# Task 1

```python
### Step 1: Predefined correct PIN and initial balance
correct_pin = "9898"
balance = 1000.00

#Step 2: Input the PIN and withdrawal amount
input_pin = input("Enter your 4-digit PIN: ")
amount_input = input("Enter withdrawal amount: ")

#Step 3: Validate amount input
try:
    amount = float(amount_input)
except valueError:
    print("Invalid Error Entered.")
    exit()

#step 4: Check if PIN is correct
if input_pin != correct_pin:
    print("Invalid PIN.")
    exit()

#Step 5: Check if the balance is sufficient
if amount > balance:
    print("Insufficient Balance.")
    exit()

#Step 6: process withdrawal
balance -= amount
print(f"Withdrawal Successful. Remaining balance: ${balance:.2f}")
```

# Task 2

```python
def is_valid_password(password):
    #Rule1: Atleeast 8 characters
    if len(password) < 8:
        return "Invalid - password must contain atleast 8 characters."

    #Flags for rules 2, 3, and 4

    has_number = False
    has_upper = False
    has_lower = False


    for char in password:
        if char.isdigit():
            has_number = True
        elif char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
            
    #final Validation
    if not has_number:
        return "Invalid - password must contain atleast one number."
    if not has_upper:
        return "Invalid - password must contain atleast one upper letter."
    if not has_lower:
        return "Invalid - password must contain atleast one lower letter."

    return "Valid password"

user_input = input("Enter your password: ")
result = is_valid_password(user_input)
print(result)
```
