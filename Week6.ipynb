# Task 2 – Class Grades Processor (.csv)

```python
import csv

#Step 1: Create sample grades.csv
grades = [
    ["Name", "Score"],
    ["Alice", 90],
     ["Musa", 88],
     ["Agnes", 83],
     ["Peter", 75],
     ["Lamb", 92],
     ["Alie", 68],
     ["John", 70],
     ["Dede", 84]
]

with open("grades.csv", "w", newline = "") as f:
    writer = csv.writer(f)
    writer.writerows(grades)
```

```python
#Step 2: Read grades and calculate pass/fail
output_rows = [["Name", "Score", "Result"]]

with open("C:/Users/hp/Desktop/notebook-demo/grades.csv", "r") as f:
    reader = csv.reader(f)
    next(reader) #Skip header
    for row in reader:
        name, score = row[0], int(row[1])
        result = "Pass" if score >= 50 else "Fail"
        output_rows.append([name, score, result])
```
